## Analysis 2

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
df.head()
df.columns
```

## Número de veículos pertencentes a cada marca
## Number of vehicles by brand

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
sns.set_style("whitegrid")
g = sns.catplot(
    data=df, kind="count",
     y="brand", 
    ci="sd", palette="Blues_r", alpha=.6, height=7, aspect=2
)
g.set_axis_labels("Count of vehicles", "Brand", fontdict={'size':16})
g.ax.set_title("Number of vehicles by brand" ,fontdict= {'size':18})
plt.show()
```

```python
# Salvando o plot
g.savefig(("plots/Analise2/brand-vehicleCount.png"))
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

## Mean vehicle price based on type and gearbox

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio
sns.set_style("whitegrid")
g = sns.catplot(
    data=df, kind="bar",
    x="vehicleType", y='price', hue="gearbox",
    ci="sd", palette="dark", alpha=.6, height=6, aspect=2
)
g.set_axis_labels("Vehicle type", "Mean price", fontdict={'size':16})
g.ax.set_title("Mean vehicle price based on type and gearbox" ,fontdict= {'size':18})
```

```python
# Salvando o plot
g.savefig("plots/Analise2/vehicletype-gearbox-price.png")
```
