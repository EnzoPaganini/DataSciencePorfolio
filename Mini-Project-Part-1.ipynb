## Analysis 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Distribuição de Veículos com base no Ano de Registro
## Vehicle distribution by year of registration 

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro
df.head()
```

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.histplot(
    df['yearOfRegistration'],
    kde=True,
    ax = ax,
    color="#ac33bb"
)
ax.set_title('Vehicle distribution by year of registration', fontsize= 15)
plt.ylabel("Density (KDE)", fontsize= 15)
plt.xlabel("Year of registration", fontsize= 15)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/vehicle-distribution.png")
```

## Variação da faixa de preço pelo tipo de veículo
## Price variation by vehicle type

```python
# Crie um Boxplot para avaliar os outliers
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(12,6))
sns.set_theme(style="ticks", palette="pastel")
sns.boxplot(x="vehicleType", y="price",
             palette=["m", "g"],
            data=df)
sns.despine(offset=10, trim=True)
ax.set_title('Outliers evaluation', fontsize= 15)
plt.ylabel("Vehicle type", fontsize= 15)
plt.xlabel("Price", fontsize= 15)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/price-vehicleType-boxplot.png")
```

## Contagem total de veículos à venda conforme o tipo de veículo
## Total of vehicles by type

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 

g = sns.catplot(
    data=df, kind="count",
    x="vehicleType",
    ci="sd", palette="dark", alpha=.6, height=6, aspect=2
)
g.set_axis_labels("Vehicle type", "Total of vehicles")
g.ax.set_title("Total of vehicles by type" ,fontdict= {'size':18})
#Loop to put the count of vehicle on top of the bar
for p in g.ax.patches:
    g.ax.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+500))
```

```python
# Salvando o plot
g.savefig("plots/Analise1/count-vehicleType.png")
```
